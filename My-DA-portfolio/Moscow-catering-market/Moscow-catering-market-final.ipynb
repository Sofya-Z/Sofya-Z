**Большая часть дубликатов в названиях удалена**

Теперь разберемся к отнесению к сети.
1. Некоторые заведения с одинаковым назваинем относятся одновременно к сетевым и несетевым. Этого надо избежать. Примем, что если из ряда заведений с одинаковым названием хоть одно относится к сетевым, для всех поставим значение `chein` "да".
2. Есть заведения, названия которых не повторяются, но они обозначены как сетевые. Оставим их в этом статусе, поскольку скорее всего сложность в названии - это скрытый дубль, котрый остался неисправленным. Другой вариант принадлежности к сети - сеть в нескольких городах, но в Москве только одно заведение.
3. Есть заведения, которые не относятся к сетевым, при этом есть несколько заведений с одинаковым названием. Это самый сложный случай для решения. Оставим их несетевыми. Причина - у составителя данных могла быть логика, которая с первого взгляда непонятна, но имеет смысл. Например, названия одинаковые, но бюджет, управление, концепция разные. Без уверенности не будем менять данные.


```python
# chain_list - список названий сетевых кафе
chain_list = df.query('chain == "да"')['object_name'].unique().tolist()

# Проверим, что для заведений из списка принадлежность к сетевым типам не всегда верна
print(df['chain'][df['object_name'].isin(chain_list)].unique())

# присвоим всем названиям заведений из этого списка значение chain "да"
df['chain'][df['object_name'].isin(chain_list)] = "да"

# Проверим, что все получилось
print(df['chain'][df['object_name'].isin(chain_list)].unique())
    
```

**С принадлежностью к сетевому типу разобрались**

Теперь необходимо все заведения с одинаковым названием отнести к одному типу.

```python
# df1 - таблица с названиями сетей и их типами заведений
df1 = df.query('chain == "да"').pivot_table(
    index='object_name', columns='object_type', values='id', aggfunc='count')

# Столбец type_max содержит название колонки с максимальным количеством заведений
# Если равное количество - возвращает первый встретившийся

df1['type_max'] = df1.idxmax(axis = 1)
df1.head()
```

```python
# Оставим только индексы и столбец type_max
df1 = df1['type_max'].reset_index()
df1
```

```python
# Теперь объеденим исходный датафрейм с нашим списком типов
# К исходному добавится столбец с типом,
# который будет одинаковый для всех заведений с одинаковым названием
df = df.merge(df1, on='object_name', how='left')

df.columns
```

```python
# Посчитаем, сколько объектов имеют тип, который не совпадает с новым типом
new_type_count = df[
    (df['type_max'] != df['object_type']) & ~df['type_max'].isna()
                    ]['id'].count()
print('Число заведений, для которых будет изменен тип данных:',
     new_type_count)
```

Для найденных 832 заведений изменим тип на тот, который встречается максимальное количество раз.

```python
df['object_type'][
    (df['type_max'] != df['object_type']) & ~df['type_max'].isna()
    ] = df['type_max'][
    (df['type_max'] != df['object_type']) & ~df['type_max'].isna()]
# Проверим, что все сработала.
# Ячейки для проверки: test_list
test_list = [16, 47, 148, 13, 25, 0, 5]
for test_index in test_list:
    display(df.loc[test_index])
```

Проверим также, сработала ли логика, повторив подсчет неуникальных типов у заведений с одинаковым названием

```python
# df1 - таблица с названиями сетей и их типами заведений
df2 = df.query('chain == "да"').pivot_table(
    index='object_name', columns='object_type', values='id', aggfunc='count')

# Колонка count - количество различных типов для каждого названия
df2['count'] = df2.count(axis=1)

df2.query('count != 1').count()
```

По данным из других источников - Яндекс.Карты, все представленные адреса оказываются московскими, по указаным адресам располоджены кафе. 
Город Московский также является административной частью Москвы.
Были проверены 13 мест из 69. Предположим, что все эти адреса - действительно кафе, просто не указан город Москва.

Тогда удалим из строк указание города Московский.

```python
# Удалим название города из строк с указанием города Московский.
# Для начала создадим отдельный столбец для названия улицы
df['street'] = df['address']

# Создадим список строк, где город Московский и циклом удалим там его название
drop_line = [8882, 8883, 8893]
for i in drop_line:
    df.loc[i, 'street'] = df.loc[i, 'street'][18:]
    display(df.loc[i, 'street'])
```

```python
# Удалим название города из строк с указанием города Москова.
# Для этого определим функцию удаления строки до первой запятой включительно
# Но только если comma_index == 12

def drop_city(row):
    if row['comma_index'] == 12:
        row['street'] = row['street'][14:]
    return row

df = df.apply(drop_city, axis=1)
df.head()
```

```python
df['street'].str.index(',')
```

Теперь положение первой запятой отделяет название улицы от дома и пр. Напишем функцию, удаляющую лишнюю информацию.

```python
# Определим drop_street для отделения отдельно улицы
def drop_street(street):
    drop_index = street.index(',')
    street = street[:drop_index]
    return street

df['street'] = df['street'].apply(drop_street)

# Удалим ненужный столбец
df = df[['id', 'object_name', 'chain', 'object_type', 'address', 'number',
       'street']]
df.head()
```

### Виды объектов, для которых характерно сетевое распространение

Подготовим таблицу для исследования соотношения типов заведений в сетях и вне их.
Кроме абсолютных значений, было бы интересно посмотреть и процентное распределение.

```python
# Таблица числа разных типов для сетевых точек
chain_type_yes = df.query(
            'chain == "да"').pivot_table(
    index=['chain', 'object_type'], values='id', aggfunc='count').reset_index()

# Таблица числа разных типов для несетевых точек
chain_type_no = df.query(
            'chain == "нет"').pivot_table(
    index=['chain', 'object_type'], values='id', aggfunc='count').reset_index()

# Добавим рассчитанный процент
chain_type_yes['share'] = chain_type_yes['id']/chain_type_yes['id'].sum()*100
chain_type_no['share'] = chain_type_no['id']/chain_type_no['id'].sum()*100

# Сольем две таблицы
chain_type = chain_type_no.append(chain_type_yes, ignore_index=True)
chain_type
```

Таблица `chain_type` содержит количество заведений для каждого типа для сетевых и несетевых точек + процент их относительно всех сетевых/несетевых заведений.

Теперь можно отобразить наглядно распределение.

```python
fig = plt.figure(figsize=(14, 4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.barplot(x='object_type', y='id', data=chain_type, hue='chain', ax=ax1)

sns.barplot(x='object_type', y='share', data=chain_type, hue='chain', ax=ax2)
sns.despine()
ax1.tick_params(axis='x', rotation=-60)
ax1.set_xlabel('Тип заведения')
ax1.set_ylabel('Число заведений')
ax1.set_title('Распределение заведений по типам')

ax2.tick_params(axis='x', rotation=-60)
ax2.set_xlabel('Тип заведения')
ax2.set_ylabel('Доля заведений, %')
ax2.set_title('Процентное соотношение по типам')
plt.show()
```

Среди самых популярных четырех типов заведений: кафе, ресторан, столовая, фастфуд, распределение сетевых и несетевых заведений сильно отличается. 
Отобразим отдельно эти четыре типа.


```python
# Напишем функцию категоризации для типов:
# Рестораны, кафе, столовые и фастфуд оставляет неизменным,
# Остальные типы объединяются в другие
def other_types(type):
    if type == "кафе":
        return type
    if type == "столовая":
        return type
    if type == "ресторан":
        return type
    if type == "фастфуд":
        return type
    else:
        return "другой тип"

# Применим ее к столбцу object_type в исходном датафрейме
df['type_group'] = df['object_type'].apply(other_types)

# Теперь создадим еще одну таблицу для анализа типов с более крупными типами

# Таблица числа разных типов для сетевых точек
chain_type_yes = df.query(
            'chain == "да"').pivot_table(
    index=['chain', 'type_group'], values='id', aggfunc='count').reset_index()

# Таблица числа разных типов для несетевых точек
chain_type_no = df.query(
            'chain == "нет"').pivot_table(
    index=['chain', 'type_group'], values='id', aggfunc='count').reset_index()

# Добавим рассчитанный процент
chain_type_yes['share'] = chain_type_yes['id']/chain_type_yes['id'].sum()*100
chain_type_no['share'] = chain_type_no['id']/chain_type_no['id'].sum()*100

# Сольем две таблицы
chain_type_group = chain_type_no.append(chain_type_yes, ignore_index=True)
chain_type_group
```

Теперь на основе обновленной таблицы типов, построим графики.

```python
fig = plt.figure(figsize=(14, 4))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.barplot(x='type_group', y='id', data=chain_type_group, hue='chain', ax=ax1)

sns.barplot(x='type_group', y='share', data=chain_type_group, hue='chain', ax=ax2)
sns.despine()
ax1.tick_params(axis='x', rotation=-60)
ax1.set_xlabel('Тип заведения')
ax1.set_ylabel('Число заведений')
ax1.set_title('Распределение заведений по типам')


ax2.tick_params(axis='x', rotation=-60)
ax2.set_xlabel('Тип заведения')
ax2.set_ylabel('Доля заведений, %')
ax2.set_title('Процентное соотношение по типам')
plt.show()
```

Теперь посмотрим, для каких типов характерна наибольшая доля сетевых заведений.

### Топ-10 улиц по количеству объектов

Создадим сводную таблицу, которая отобразит число заведений на различных улицах.

```python
df.pivot_table(
    index='street', values='id', aggfunc='count').sort_values(by='id', ascending=False).head(10)
```

Формирование адреса в городе Зеленограде происходит не по обычным правилам - иногда дается название улиц, иногда только номер корпуса. Исключим его из анализа. Также исключим и поселение Сосенское - как неинтересное для открытия нашего заведения.

```python
top_street = df.query(
    'street != "город Зеленоград" and street != "поселение Сосенское"').pivot_table(
    index='street', values='id', aggfunc='count').sort_values(by='id', ascending=False
    ).head(10).reset_index()
top_street
```

Отобразим данные на диаграме

```python
sns.barplot(x='street', y='id', data=top_street, color='mediumseagreen')
sns.despine()
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Число заведений')
plt.title('Топ-10 улиц')
plt.show()
```

```python
print('На топ 10 улицах расположено заведений, %:', round(top_street['id'].sum()*100/len(df), 2))
```

Выведем районы Москвы, где расположены эти улицы.

```python
top_street_list = top_street["street"]
top_district = df.query(
    'street in @top_street_list')['District'].unique()
print(top_district)
print()
print('Всего районов, через которые проходят эти улицы:', 
     len(top_district)-1)
```

Большинство улиц с максимальным числом заведений очень протяженные, проходят через несколько районов. Поэтому список районов с найденными улицами оказался столько внушительным.

Поэтому выведем отдельно топ-10 районов с максимальным числом заведений.

```python
top_district = df.pivot_table(
    index='District', values='id', aggfunc='count'
).sort_values(by='id', ascending=False).head(10).reset_index()
top_district
```

```python
print('В топ 10 районах расположено заведений, %:', round(top_district['id'].sum()*100/len(df), 2))
print('В топ 3 районах расположено заведений, %:', round(top_district.loc[0:2, 'id'].sum()*100/len(df), 2))
```

```python
sns.barplot(x='District', y='id', data=top_district, color='mediumseagreen')
sns.despine()
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Число заведений')
plt.title('Топ-10 районов')
plt.show()
```

### Улицы с одним заведением

Выведем список улиц с одним заведением на них.

```python
bottom_street = df.pivot_table(
    index='street', values='id', aggfunc='count').sort_values(by='id', ascending=True
    ).reset_index().query(
    'id == 1')
bottom_street['id'].count()
```

```python
print('На этих улицах расположено заведений, %:', round(bottom_street['id'].sum()*100/len(df), 2))
```

Таких улиц ожидаемо оказалось много - более 552. Найдем районы, в которых улицы расположены.

```python
bottom_street_list = bottom_street["street"]
bottom_district = df.query(
    'street in @bottom_street_list')['District'].unique()
print(bottom_district)
print()
print('Всего районов, через которые проходят эти улицы:', 
     len(bottom_district)-1)
```

В 95 районах Москвы есть улицы с только одним заведением.

Посмотрим, есть ли улицы только с одним районом из топового района. Для расположения нового кафе возможно именно такое положение может быть самым выгодным - район популярный, много заведений, значит поток покупателей достаточно большой, при этом в на той же улице будет мало конкурентов.

```python
top_district_list = top_district['District']
len(df.query(
    'street in @bottom_street_list and District in @top_district_list')
    ['street'].unique())
```

Тоже много. Попробуем ограничить только топ-3 района.

```python
top_3_district_list = ['Тверской район', 'Пресненский район', 'Басманный район']
len(df.query(
    'street in @bottom_street_list and District in @top_3_district_list')
    ['street'].unique())
```

Отлично, значит в этих трех районах есть 50 улиц, где будет небольшая конкуренция.
