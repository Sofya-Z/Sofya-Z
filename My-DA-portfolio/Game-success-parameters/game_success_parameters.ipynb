### Форматы данных

Приведем типы столбцов к нужному типу.
1. Год преобразовать в `int`
2. Рейтинг пользователей преобразовать в числовой тип

```python
df = df.dropna(subset=['year_of_release'])

# Приведем год к типу int
df['year_of_release'] = df['year_of_release'].astype('int')
df.info()
```

Все получилось.

### Проверка наличия дубликатов и аномальных значений

В первую очередь проверим наличие аномальных значений в столбцах с типом `str`.

```python
object_columns = ['name', 'platform', 'genre', 'rating']
for col in object_columns:
    print(df[col].unique())
```

Заметных аномалий нет.

Теперь проверим на аномальность числовые значения.

```python
float_columns=['year_of_release', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'critic_score', 'user_score']
for col in float_columns:
    print('Данные по параметру', col)
    print()
    print(df[col].describe())
    df[col].plot(kind='box', grid=True, title='Распределение значений признака')
    plt.show()
```

Данные по продажам копий и продажам в разных регионах имеют сильно ассиметричное распределение - игры, копий которых было продано аномально много.

Для того, что бы выбросы не влияли на сделанные в ходе исследования выводы, удалим их.

Посчитаем, сколько игр попадает под разные критерии отсеивания.

```python
# Список колонок с данными по продажам
drop_columns=['na_sales', 'eu_sales',
       'jp_sales', 'other_sales']
# Выведем значения 99 и 95 перцентилей для этих колонок
for col in drop_columns:
    print('Параметр', col)
    print()
    percent = np.percentile(df[col], 99)
    print('99-й перцентиль:', percent)
    print('Количество значений выше 99-го перцентиля:', len(df[df[col] > percent]))

    print()
    percent = np.percentile(df[col], 95)
    print('95-й перцентиль:', percent)
    print('Количество значений выше 95-го перцентиля:', len(df[df[col] > percent]))
```

Выберем границей нормальности 99-й перцентиль и удалим все данные, выше этих значений для параметра продаж.

```python
# Посчитаем количество удаленных фильмов
deleted = 0

for col in drop_columns:
    percent = np.percentile(df[col], 99)
    deleted += len(df[df[col] > percent])
    df = df[df[col] <= percent]
print(f'Всего удалено {deleted} игр')
print(f'Что составило {deleted/len_df: .2%} от исходного датафрейма')
df.info()
```

Всего с аномально высокими продажами в разных регионах было удалено 642 игры - 3.8% от исходного датафрейма.

### Выходы игр на различных платформах

1. Изучим наиболее популярные платформы.
2. Посмотрим на распредедение по годам выхода игр на наиболее популярных платформах

Изучим популярность платформ с точки зрения количества выходящих игр и суммарных продаж.

```python
# Построим распределение по платформам в двух вариантах

fig = plt.figure(figsize=(14,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

df.groupby('platform')['name'].agg('count').sort_values(ascending=False).plot(kind='bar', ax=ax1)
ax1.set_title('Выход игр по платформам')
ax1.set_ylabel('Количество игр')
ax1.set_xlabel('Платформа')

df.groupby('platform')['total_sales'].agg('mean').sort_values(ascending=False).plot(kind='bar', ax=ax2)
ax2.set_title('Средние продажи по платформам')
ax2.set_ylabel('Количество игр')
ax2.set_xlabel('Платформа')
plt.show()

```

Топ платформ с точки зрения суммарных продаж и количества игр очень отличаются. Понятно, что количество игр и их продажи на разных платформах сильно отличаются от года к году. ### Связь продаж с отзывами

Посмотрим, как влияют на продажи отзывы пользователей и критиков.

Для начала проанализируем продажи на одной самой популярной `X360`.

```python
# Напишем функцию для визуализации продаж с оценками для определенной платформы
def corr_sales_scores(platf):
    # Формируем вспомогательный датасет
    df_platf = df_2012.query('platform==@platf')
    # Выводим коэффициенты корреляции
    print(f"Корреляция между суммарными продажами на платформе {platf} и оценкой критиков:",
         round(df_platf['total_sales'].corr(df_platf['critic_score']), 3))
    print(f'Корреляция между суммарными продажами на платформе {platf} и оценкой пользователей:',
         round(df_platf['total_sales'].corr(df_platf['user_score']), 3))
        
    # Рисуем два графика связи продаж с оценками
    fig = plt.figure(figsize=(15, 5))
    ax1 = fig.add_subplot(1,2,1)
    sns.regplot(data=df_platf, x="critic_score", y="total_sales", ax=ax1)

    ax1.set_xlabel('Оценка критиков')
    ax1.set_ylabel('Продажи по всем регионам')
    ax1.set_title(f'Связь продаж на платформе {platf} с оценкой критиков')
    
    ax2 = fig.add_subplot(1, 2, 2)
    sns.regplot(data=df_platf, x="user_score", y="total_sales", ax=ax2)
    
    ax2.set_xlabel('Оценка пользователей')
    ax2.set_ylabel('Продажи по всем регионам')
    ax2.set_title(f'Связь продаж на платформе {platf} с оценкой пользователей')
        
corr_sales_scores('X360')
```

По коэффициенту корреляции видно, что связь с оценкой критиков значительная - чем выше оценка критиков, тем лучше продажи.
Связи с оценкой пользователей практически нет.

Проверим наличие такой связи для остальных платформ.

```python
platform_list = ['WiiU', 'XOne', 'X360', '3DS', 'PS4', 'PS3', 'PSV', 'PC']

for platf in platform_list:
    corr_sales_scores(platf)
```

#### Вывод

1. Связь продаж с оценкой игры ниже среднего.
2. Связь продаж с оценкой критиков положительная и для большинства платформ более тесная, чем с оценкой пользователей.
3. На платформах `Wiiu`, `XOne`, `X360` и `PS3` теснота связи между уровнем продаж и оценкой критиков наиболее сильная (коэфф.корр. около 0.4).
4. Только на платформах `WiiU` и в меньшей степени `3DS` наблюдается существенная связь с оценкой пользователей.

## Гипотезы о различии пользовательских рейтингов

### Пользовательские рейтинги платформ Xbox One и PC

Проверим гипотезу о равенстве рейтингов пользователей на двух платформах.
Сначала выведем отдельными датафреймами рейтинги игр на двух платформах и их распределение.

```python
# Датафреймы с рейтингами платформ XOne и PC
# Удалим строки с пропусками для корректного анализа данных
df_xone = df_2012.query('platform == "XOne"').dropna()
df_pc = df_2012.query('platform == "PC"').dropna()

fig = plt.figure(figsize=(14,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.histplot(data=df_xone, x='user_score', bins=20, kde=True, ax=ax1)
ax1.set_xlabel('Оценка пользователей')
ax1.set_ylabel('Число игр')
ax1.set_title('Оценка пользователей на платформе Xbox One')

sns.histplot(data=df_pc, x='user_score', bins=20, kde=True, ax=ax2)
ax2.set_xlabel('Оценка пользователей')
ax2.set_ylabel('Число игр')
ax2.set_title('Оценка пользователей на платформе PC')

plt.show()
```

Распределение оценок имеет левостороннюю ассиметрию и далеко от нормального. Проверим критерием Шапиро_Уилка нормальность распределения двух выборок.

Н0: Выборка распределена по нормальному закону.
Н1: Распределение другое.

Выберем уровень значимости 0.05

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(df_xone['user_score'])
p_value = results[1]

print('Проверка нормальности расределения оценок пользователя на платформе XboxOne')
print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
    
    
results = st.shapiro(df_pc['user_score'])
p_value = results[1]

print()
print('Проверка нормальности расределения оценок пользователя на платформе PC')
print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
```

Критерий проверки показал, что данные распределены не по нормальному закону. Тогда для проверки гипотезы о равенстве средних значений воспользуемся непараметрическим критерием Манна-Уитни.

Н0: Средние значения в двух выборках равны.
Н1: Средние значения различаются.

Оставим уровень значимости 0.05

```python
# Выведем значения распределений
print('Статистики пользовательской оценки на платформе XboxOne')
print(df_xone['user_score'].describe())
print()
print('Статистики пользовательской оценки на платформе PC')
print(df_pc['user_score'].describe())
print()

# Расчитаем значение критерия
results = st.mannwhitneyu(df_xone['user_score'], df_pc['user_score'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

#### Вывод 

Гипотеза о равенстве средних значений пользовательских оценок на платформах XboxOne и PC не отвергается после статистической проверки.
