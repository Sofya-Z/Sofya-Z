# Прогнозирование оттока клиентов

## Описание проекта

Сеть фитнес-центров разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Чтобы бороться с оттоком, необходимо понять, какие признаки связаны с вероятным оттоком клиентов, можно ли это спрогнозировать.

**Цель:** провести анализ оттока клиентов и подготовить план действий по их удержанию.


**Задачи:**
1. Научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
2. Сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
3. Проанализировать основные признаки, наиболее сильно влияющие на отток;
4. Разработать рекомендации по повышению качества работы с клиентами:
 * выделить целевые группы клиентов;
 * предложить меры по снижению оттока;
 * определить другие особенности взаимодействия с клиентами.

В нашем распоряжении имеются цифровые анкеты клиентов со следующей информацией:
* данные социальной анкеты (пол, возраст, откуда пришел в клуб и прочее);
* информация о посещении и текущем статусе абонемента;
* факт оттока/активности клиента.

## EDA (исследовательский анализ данных)

Изучим распределение количественных признаков.

Для этого напишем автоматизирующую процесс функцию.

```python
def features_distribution(feature):
    fig = plt.figure(figsize=(15,6))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    # НА первом графике - гистограмма распределения
    sns.histplot(data=df, x=feature, ax=ax1)
    ax1.set_title('Распределение признака')
    
    # На втором графике - сравнение распределения в двух целевых подгруппах
    sns.kdeplot(data=df
            , x=feature
            , bw_adjust=0.5
            , hue="Churn",
            ax=ax2)
    ax2.set_title('Распределение признака по группам')
    plt.show()
    
    # Статистический тест Манна-Уитни на равенство распределений
    p_lavel = st.mannwhitneyu(
          df.query('Churn == 0')[feature],
          df.query('Churn == 1')[feature])[1]
    alpha = 0.05
    if p_lavel < alpha:
        print(f'По признаку {feature} группы различаются')
        print('Уровень различий равен', round(p_lavel, 4))
    else:
        print(f'По признаку {feature} различие между группами не подтверждается')
        print('Уровень различий равен', round(p_lavel, 4))
```

Реализуем функция для количественных признаков.

Признаки:

* `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
* `Age` - возраст;
* `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
* `Lifetime` - количество месяцев с первого обращения клиента в клуб;
* `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
* `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
* `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

```python
features_list = ['Contract_period', 'Age', 'Month_to_end_contract', 'Lifetime',
       'Avg_class_frequency_total', 'Avg_class_frequency_current_month', 'Avg_additional_charges_total']

for feature in features_list:
    features_distribution(feature)
```

## Модель прогнозирования оттока клиентов

### Разбиение выборки на обучающую и валидационную

В нашем случае прогнозирование не связано с временными рядами, значения признаков для разных клиентов не связано. Значит применим случайное разделение на две выборки. Разделим в классическом соотношении 0.8:0.2.

```python
# Разделим данные на признаки (матрица X) и целевую переменную (y)
X = df.drop('Churn', axis = 1)
y = df['Churn']

# Разделим данные на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Стандартизируем значения признаков.

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
X_train_st = scaler.fit_transform(
    X_train
)  # обучаем scaler и одновременно трансформируем матрицу для обучающей выборки

# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

### Выбор метрик

Цель заказчика - борьба с оттоком клиентов. Поэтому приоритетная задача модели - не упустить клиентов, которые быстро покинут клуб. Модель, отвечающая на вопрос "попадет ли в отток клиент?" должна редко отвечать, что не попадет про того клиента, который покинет клуб. Т.е. важно минимизировать число ложноотрицательных присвоений.

Поэтому приоритетной метрикой выберем `recall` - полноту.

Дополнительно вычислим `accuracy` - точность и `precision` - долю правильных ответов.

### Выбор модели

Используем три модели для обучения и сравним их по выбранным метрикам:

- Логистическая регрессия -	`linear_model.LogisticRegression()`
- Случайный лес - `ensemble.RandomForestClassifier()`
- Градиентный бустинг - `ensemble.GradientBoostingClassifier()`

### Обучение моделей

Для начала, получим значения метрик на трех моделях с параметрами по умолчанию.

```python
# список моделей
models = [
    LogisticRegression(random_state=0),
    RandomForestClassifier(n_estimators = 100, random_state=0),
    GradientBoostingClassifier(n_estimators = 100, random_state=0)
]

# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train_st, y_train, X_test_st, y_test):
    model = m
    model.fit(X_train_st, y_train)
    y_pred = model.predict(X_test_st)
    
    print('Recall: {:.2f}'.format(recall_score(y_test, y_pred)))
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
    print('Precision: {:.2f}'.format(precision_score(y_test, y_pred)))
    print('F1: {:.2f}'.format(f1_score(y_test, y_pred)))


# Выводим метрики по списку моделей
for i in models:
    print(f'Метрики для модели {i}')
    make_prediction(i, X_train_st, y_train, X_test_st, y_test)
    print()
```

Начнем с логичтической регрессии. Будем варьировать параметр регуляризации C.

```python
# Создадим лист перебора значений параметров
C_list = [1e-3, 1e-2, 1e-1, 0.5, 1, 2, 5, 10, 50, 100]
recall_lg_diff_C = []
accuracy_lg_diff_C = []
f1_lg_diff_C = []

# Реализуем предсказание с разными параметрами
for C in C_list:
    model = LogisticRegression(C=C, random_state=0)
    model.fit(X_train_st, y_train)
    y_pred = model.predict(X_test_st)
    
    print(f'Метрики логистической регрессии при С={C}')
    
    print('Recall: {:.2f}'.format(recall_score(y_test, y_pred)))
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
    print('Precision: {:.2f}'.format(precision_score(y_test, y_pred)))
    print('F1: {:.2f}'.format(f1_score(y_test, y_pred)))
    
    recall_lg_diff_C.append(recall_score(y_test, y_pred))
    accuracy_lg_diff_C.append(accuracy_score(y_test, y_pred))
    f1_lg_diff_C.append(f1_score(y_test, y_pred))

# Визуализируем значение метрик
fig = plt.figure(figsize=(10, 6))
plt.plot(C_list, recall_lg_diff_C, color='red', linestyle='-', marker='o', label='Recall')
plt.plot(C_list, accuracy_lg_diff_C, color='blue', linestyle='--', marker='^', label='Accuracy')
plt.plot(C_list, f1_lg_diff_C, color='green', linestyle='-.', marker='s', label='F1')
plt.xscale('log')
plt.title('Изменение метрик логистической регрессии')
plt.xlabel('Значение параметра С, логарифмическое')
plt.ylabel('Значение метрик')
plt.legend()
plt.grid()
plt.show()
```

**Вывод:** выберем значение параметра `C = 2` - наименьшее значение из тех, когда качество предсказания достигло своего максимума. Выберем наименьшее для того, что бы не переобучить модель.

Сделаем проверку гипотезы о равенстве долей ушедших клиентов в различных кластерах. Проверку будем делать при помощи z-статистики, сравнение будет проходить попарно.

```python
# Для удобства, отберем только интересующие признаки
cluster_churn=df[['cluster_km', 'Churn']]
```

```python
# Напишем функцию, которая посчитает p-уровень - вероятность ошибки при равенстве долей в двух кластерах
def p_level_2_clusters(cl_1, cl_2):
    churn_1 = cluster_churn.query('cluster_km == @cl_1')['Churn'].sum()
    churn_2 = cluster_churn.query('cluster_km == @cl_2')['Churn'].sum()
    amount_1 = cluster_churn.query('cluster_km == @cl_1')['Churn'].count()
    amount_2 = cluster_churn.query('cluster_km == @cl_2')['Churn'].count()
    
    # пропорция оттока в первом кластере
    p1 = churn_1/amount_1
    # пропорция успехов во втором кластере
    p2 = churn_2/amount_2

    # пропорция успехов в комбинированном датасете:
    p_combined = (churn_1 + churn_2) / (amount_1 + amount_2)
    # разница пропорций в кластерах
    difference = p1 - p2
    
    # считаем статистику в ст.отклонениях стандартного нормального распределения
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/amount_1 + 1/amount_2))
    # задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
    distr = stats.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    if p_value >= alpha:
        print(f'Сравнение долей оттока клиентов в классах {cl_1} и {cl_2}')
        print('p-значение: ', p_value)
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )
        
        print()
    return p_value
```

```python
p_data = []

alpha = 0.05
for cl_1 in df['cluster_km'].unique():
    for cl_2 in df['cluster_km'].unique():
        if cl_1 >= cl_2:
            continue
        
        p = p_level_2_clusters(cl_1, cl_2)
        p_data.append([cl_1, cl_2, p])
        

df_p = pd.DataFrame.from_records(p_data)
df_p.columns = ['Класс_1', 'Класс_2', 'p-уровень']
df_p.head(10)
```

Предположение подтвердилось - доли оттока клиентов в парах кластеров 2-4 и 1-3 не различаются. Видимо, выделение произошло по независящим параметрам.

Анализ средних значений признаков показывает, что принципиальное различие между классами 2-4 и 1-3 по полу клиентов. А класс 0 выделяется от других отсутствием информации о номере телефона.
